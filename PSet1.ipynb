# Prompt
_3. Data cleaning: "data_students.tsv" contains a dataset with information about students. Write the appropriate code for each of the tasks below. This section should be completed without generative AI, but with open access to static online resources like course material, stackoverflow, blogs with sample code, etc. # Import "data_students.tsv" into a dataframe # Rename the columns to improve machine and human-readability # Display the number of missing values in each column # Create a new column identifying whether the age column is missing # Replace missing values in the 'age' column with the mean age # Convert the 'gender' column to lowercase # Split the location column into separate columns for city and state # Save the cleaned data to a new CSV file with an appropriate name_

```python
# 0) Setup
import pandas as pd
from pathlib import Path

# 1) Import "data_students.tsv" into a dataframe
df = pd.read_csv("data_students.tsv", sep="\t")

# 2) Rename columns to improve machine & human readability
#    - lowercase, strip, replace non-alphanumerics with underscores, collapse repeats, trim edges
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(r"[^0-9a-zA-Z]+", "_", regex=True)
      .str.replace(r"_+", "_", regex=True)
      .str.strip("_")
)

# 3) Display the number of missing values in each column
print("Missing values per column:")
print(df.isna().sum())

# 4) Create a new column identifying whether the age column is missing
#    (coerce to numeric first in case age was read as string)
if "age" in df.columns:
    df["age"] = pd.to_numeric(df["age"], errors="coerce")
    df["age_missing"] = df["age"].isna()
else:
    # If the dataset used a different header (e.g., 'student_age'), adapt as needed.
    raise KeyError("Expected an 'age' column after renaming.")

# 5) Replace missing values in the 'age' column with the mean age
age_mean = df["age"].mean(skipna=True)
df["age"] = df["age"].fillna(age_mean)

# 6) Convert the 'gender' column to lowercase (if present)
if "gender" in df.columns:
    df["gender"] = df["gender"].astype("string").str.strip().str.lower()

# 7) Split the 'location' column into separate 'city' and 'state' columns (if present)
#    Assumes format like "Nashville, TN" or "Chicago, Illinois"
if "location" in df.columns:
    # Split on the first comma only
    loc_split = df["location"].astype("string").str.split(",", n=1, expand=True)
    df["city"] = loc_split[0].str.strip()
    df["state"] = (
        loc_split[1].str.strip() if loc_split.shape[1] > 1 else pd.NA
    )

# 8) Save the cleaned data to a new CSV file
out_path = Path("data_students_clean.csv")
df.to_csv(out_path, index=False)
print(f"Cleaned file saved to: {out_path.resolve()}")
```
